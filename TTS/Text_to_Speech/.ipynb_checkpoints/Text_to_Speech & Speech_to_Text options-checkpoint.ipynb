# Text to Speech

### pyttsx3
https://github.com/nateshmbhat/pyttsx3

```python
#pip install pyttsx3
```

The installation was fast and straightforward.

This library works 100% offline with synthesized voices.

```python
import pyttsx3
engine = pyttsx3.init()
```

```python
%%time
engine.say("I will speak this text")
engine.runAndWait()
```

Very simple and understandable
